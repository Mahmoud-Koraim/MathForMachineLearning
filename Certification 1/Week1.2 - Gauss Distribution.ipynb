```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, sigma=3, mu=3):
    sigma2Pi = sigma * np.sqrt(2 * np.pi)
    xMuSquared = np.power(x - mu, 2)
    Sigma2Squared = (2 * sigma * sigma)
    return (1 / sigma2Pi ) *\
        np.exp(- xMuSquared / Sigma2Squared)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, gauss(x))
axes = plt.gca()
axes.set_xlim([0,10])
axes.set_ylim([0,.5])
plt.title('Gauss Distribution')
plt.show()
```
